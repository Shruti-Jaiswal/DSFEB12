```python
# importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # mathematical plotting lib. and pyplot - python plot 
```

```python
dataset = pd.read_csv("/content/Mall_Customers.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
# Indepenent and dependent variables (depend on problem statement)
x = dataset.iloc[:,[3,4]].values
```

```python
# Fitting the model
# Find the optimal number of cluster
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x,method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
# Fitting the Hierarchical Clustering to the dataset
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(x)
```

```python
y_hc
```

```python
# Visualising the cluster
plt.scatter(x[y_hc == 0, 0], x[y_hc == 0, 1], s = 100, c = 'red', label ='Cluster 1')
plt.scatter(x[y_hc == 1, 0], x[y_hc == 1, 1], s = 100, c = 'green', label ='Cluster 2')
plt.scatter(x[y_hc == 2, 0], x[y_hc == 2, 1], s = 100, c = 'blue', label ='Cluster 3')
plt.scatter(x[y_hc == 3, 0], x[y_hc == 3, 1], s = 100, c = 'yellow', label ='Cluster 4')
plt.scatter(x[y_hc == 4, 0], x[y_hc == 4, 1], s = 100, c = 'magenta', label ='Cluster 5')
plt.title("Cluster of Customers")
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
# Visualising the cluster
plt.scatter(x[y_hc == 0, 0], x[y_hc == 0, 1], s = 100, c = 'red', label ='Stingy')
plt.scatter(x[y_hc == 1, 0], x[y_hc == 1, 1], s = 100, c = 'green', label ='Normal')
plt.scatter(x[y_hc == 2, 0], x[y_hc == 2, 1], s = 100, c = 'blue', label ='Target')
plt.scatter(x[y_hc == 3, 0], x[y_hc == 3, 1], s = 100, c = 'yellow', label ='Shopaholic')
plt.scatter(x[y_hc == 4, 0], x[y_hc == 4, 1], s = 100, c = 'magenta', label ='Poor')
plt.title("Cluster of Customers")
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
